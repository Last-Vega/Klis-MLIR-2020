## Task-1 on Gradient Descend Method

**Gradient Descend Method** is commonly used to get the optimal parameters that minimize the cost function. Regarding this method, please try to answer the following questions based on your own understanding:



### Q-1: explain how to use Gradient Descend Method for optimization;

Your answer: 

First, define the parameter as 'w' and loss function as 'L'. 
$$
y = w^tx, L = \frac{1}{n} \sum(y_i - w^tx_i)
$$
Then differentiate the loss function and find the value at which the loss function is minimized. 
$$
w_n = \frac{dw}{dL},  w_{n+1} = w_n - ε \frac{dw}{dL} w_n
$$
After that update the parameter in the direction of decreasing the loss. And repeat this flow.
