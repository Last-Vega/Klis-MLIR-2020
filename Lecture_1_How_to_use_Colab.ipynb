## 【重要】セル内変数の同期について

ここまでは、各セルの実行を独立したものとして説明してきました。しかし、実際はノートブック内の変数は同期しています。このような変数のことを**グローバル変数**と呼びます。`Colab Notebook`で作成・操作する変数は、どのセルに属するものであれ全てこのグローバル変数であることを覚えておいて下さい。

変数の同期について理解するために、３つのコードセルを以下の順番で実行して、その結果を確認しましょう。

> hensu = 10  
> print(hensu)  
> hensu = hensu + 10  
> print(hensu)  
> hensu = hensu + 10  
> print(hensu)

````python
hensu = 10
````

````python
print(hensu)
````

````python
hensu = hensu + 10
````

---

- `print(hensu)`行の1回目の実行結果は`10`
- `print(hensu)`行の2回目の実行結果は`20`
- `print(hensu)`行の3回目の実行結果は`30`

という結果になったと思います。

すなわち、３つのセルの`hensu`変数は同期しており、**前のセルの実行結果が、次のセルの実行結果に影響を与える**ということです。これは非常に重要なことですので、必ず覚えておいてください。

### ノートブックを再開する時は・・・
演習中にノートブックを途中で閉じて、別の機会に再開することがあると思います。その場合は、変数への操作を時系列に捕捉できるように、**ノートブックの先頭からコードセルを順番に再実行**するように心がけましょう。また、なるべく一つのセルの中だけで、全ての変数の初期化や処理が完了するように心がけましょう。

## Markdown記法

`Colab Notebook`のテキストセルは、`Markdown`という記法を使って書くことができます。`Markdown`記法では比較的単純な記号を使うことで、見出しや箇条書きなどを`Colab Notebook`上に記述することができます。`Wiki`記法をご存知の方は、それに似たものだと思ってください。

本演習では、みなさんにこの`Markdown`記法を使って、作成するプログラムの説明文を書いていただきます。また、必要に応じて自分自身のメモ書きを、`Notebook`内の好きな場所に残すことも可能です。

### テキストセルの作成
`Colab Notebook`の上部メニューアイコンにある`＋ テキスト`をクリックすると、新しいテキストセルが現在位置の下に作成されます。実際にテキストセルを追加すると、テキストセルの入力欄とその表示プレビューが表示されます。テキストセルの外側をクリックするとそのテキストセルは実行されます。

### 見出し
見出しを作成するには、見出し語の前に`#`記号をつけます。`#`と見出し語の間には**空白文字**が必要です。

- `#` が大見出し
- `##` が中見出し
- `###` が小見出し

という要領です。大見出しはかなり大きく文書タイトルに使われることが多いです。  
普段のメモは、**中見出しか小見出しが適切**だと思います。

### 箇条書き
上の例にもありますが、箇条書きは、先頭に`-`記号をつけます。

- 番号なし箇条書き
- 番号なし箇条書き

### 番号付き箇条書き
番号付きの箇条書きは、先頭に`1.`を書きます。いくつアイテムを羅列しても、**1.を使い続ける**のがポイントです。

```
1. 番号付き箇条書き
1. 番号付き箇条書き
```

と`Markdown`記法で書くと

1. 番号付き箇条書き
1. 番号付き箇条書き

のように**自動的に番号が調整**されて変換されます。したがってアイテムの順番を変えても、番号を振り直す必要がなく便利です。ただし、`1.`のあとに空白文字が必要ですので、注意してください。

### 引用
字下げされた引用ブロックを書くには、先頭に`>`記号を書きます。

```
> 我輩は猫である
```

と`Markdown`記法で書くと

> 我輩は猫である

と表示されます。

複数行の引用ブロックを書くには、本セル末尾の改行に関する説明を参考にしてください。改行は引用ブロックでも同じ作法です。


### 強調
ある文字列を強調するには、強調したい文字列を`**`記号で囲みます。

```
**1.を使い続ける**のがポイントです。
```

と`Markdown`記法で書くと

**1.を使い続ける**のがポイントです。

と表示されます。

### キーワード表現
`Markdown`記法を使った文書では、`Markdown` ←のように変数名、コマンド名、その他のキーワードをハイライトする表現がよく使われます。これを実現するには、キーワード表現したい文字列を ``` 記号（バッククォート）で囲みます。

```
`Markdown`記法を使った文書
```

と`Markdown`記法で書くと

`Markdown`記法を使った文書

と表示されます。

### 区切り線
区切り線は、`---`を使うと現れます。

### 改行
`Markdown`記法では、基本的に改行の概念がありません。編集モードで改行しても閲覧モードでは改行なしの文章として表現されます。強制的に改行を入れたい場合は、改行を入れたい文の末尾に**半角空白文字を二つ**挿入します。これはぱっと見てもわからないので、注意が必要です。

```
文章1□□（半角空白文字二つ）
文章2
```

と書くと

文章1  
文章2

と改行されます。

### その他
`Markdown`記法では、これ以外にも表・画像の貼り付け・ハイパーリンクなどを記述することも可能です。  
詳しくは、[参考文献](#参考文献)を見てください。
